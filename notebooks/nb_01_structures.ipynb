## Lien avec Drone Rescue

Dans le jeu, **chaque entité** (drone, tempête, survivant) est d'abord
représentée par un dictionnaire avant qu'on apprenne les classes (module 05).

**Convention de coordonnées du projet :**
- `colonne` : lettre str `'A'`…`'L'` (12 colonnes)
- `ligne` : entier 1-based `1`…`12` (12 lignes)
- La grille est une liste de listes : `grille[ligne-1][ord(col)-ord('A')]`


````python
# Exemple Drone Rescue — grille 3×3 simplifiée
grille = [
    ['.', 'B', '.'],   # ligne 1
    ['.', '.', 'S'],   # ligne 2
    ['H', '.', '.']    # ligne 3
]

# Accéder à la case colonne 'B', ligne 1
colonne, ligne = 'B', 1
j = ord(colonne) - ord('A')   # index 0-based de la colonne
i = ligne - 1                 # index 0-based de la ligne
print(f'Case {colonne}{ligne} : {grille[i][j]}')  # B

# Etat initial de la partie
etat_jeu = {'tour': 1, 'score': 0, 'survivants_restants': 10, 'partie_terminee': False}
print(f'Tour {etat_jeu["tour"]} — score : {etat_jeu["score"]} — survivants : {etat_jeu["survivants_restants"]}')
````
