### Dictionnaires
```python
drone = {'id': 'D1', 'col': 1, 'lig': 2, 'batterie': 10}
```

### Tuples (coordonnées immutables)
```python

```


````python
# Exemples exécutables — types de base
batterie_max  = 20
batterie_init = 10
batterie_courante = batterie_init - 3
print(f'Batterie : {batterie_courante}/{batterie_max}')

# Liste des identifiants de drones
ids_drones = ['D1', 'D2', 'D3', 'D4', 'D5', 'D6']
ids_drones.append('D7')
ids_drones.remove('D7')
print(f'{len(ids_drones)} drones actifs : {ids_drones}')

# Dictionnaire drone — convention 0-based (col=1 → colonne B, lig=2 → ligne 3)
LETTRES = list('ABCDEFGHIJKL')
drone = {'id': 'D1', 'col': 1, 'lig': 2, 'batterie': 10, 'survivant': None}
drone['batterie'] -= 1
drone['col'] = 2   # avance d'une colonne
print(drone)
# Affichage lisible : col=2 → 'C', lig=2 → ligne 3
print(f"Position : {LETTRES[drone['col']]}{drone['lig'] + 1}")
````

## Lien avec Drone Rescue

Dans le jeu, **chaque entité** (drone, tempête, survivant) est un dictionnaire.

**Convention de coordonnées du projet (0-based entiers) :**
- `col` : entier `0`…`11` — s'affiche `A`…`L` via `LETTRES[col]`
- `lig` : entier `0`…`11` — s'affiche `1`…`12` via `lig + 1`
- La grille est indexée `grille[lig][col]`
- Un tuple de position s'écrit `(col, lig)` — colonne en premier

> **Note pédagogique** : dans les exercices de ce module, on utilise parfois
> des lettres et des numéros 1-based pour faciliter la lecture (ex: `'B3'`).
> Le projet réel stocke tout en entiers 0-based et convertit à l'affichage.


````python
# Exemple Drone Rescue — grille 3×3 simplifiée (0-based)
grille = [
    ['.', 'B', '.'],   # lig=0 (ligne 1)
    ['.', '.', 'S'],   # lig=1 (ligne 2)
    ['H', '.', '.']    # lig=2 (ligne 3)
]

# Accéder à la case col=1, lig=0  → affichage 'B1'
col, lig = 1, 0
print(f'Case B1 : {grille[lig][col]}')  # B

# État initial de la partie
etat_jeu = {'tour': 1, 'score': 0, 'partie_finie': False}
print(f"Tour {etat_jeu['tour']} — score : {etat_jeu['score']}")
````
