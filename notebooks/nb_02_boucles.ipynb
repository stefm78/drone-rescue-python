## Concepts clés

### if / elif / else
```python
if batterie == 0:
    print('Drone HS')
elif batterie < 5:
    print('Batterie faible')
else:
    print('Drone opérationnel')
```

### for sur un dict
```python
for did, drone in drones.items():
    print(did, drone['batterie'])

# Avec index : enumerate()
for i, did in enumerate(drones):
    print(i, did)
```

### while avec break
```python
while True:
    commande = input('> ').strip().upper()
    if commande in ('NEXT', 'QUIT'):
        break
    print('Commande non reconnue')
```

### any() / all()
```python
tous_sauves = all(s['etat'] == 'sauve' for s in survivants.values())
un_hs       = any(d['hors_service'] for d in drones.values())
```


````python
# Exemples exécutables — drones comme dict de dicts (convention projet)
drones = {
    'D1': {'id': 'D1', 'batterie': 10, 'hors_service': False},
    'D2': {'id': 'D2', 'batterie': 0,  'hors_service': True},
    'D3': {'id': 'D3', 'batterie': 3,  'hors_service': False},
}

# for + if
for did, d in drones.items():
    if d['hors_service']:
        print(f'{did} : HS')
    elif d['batterie'] < 5:
        print(f'{did} : batterie faible ({d["batterie"]})')
    else:
        print(f'{did} : OK ({d["batterie"]})')

# any() / all()
print(f'Un HS ? {any(d["hors_service"] for d in drones.values())}')
print(f'Tous OK ? {all(not d["hors_service"] for d in drones.values())}')
````

## Lien avec Drone Rescue

La **boucle principale du jeu** est un `while not etat['partie_finie']` dans `console.py`.
Chaque tour, on parcourt `etat['drones'].values()` avec `for` pour appliquer les règles.
La détection de fin de partie utilise `all(s['etat'] == 'sauve' for s in survivants.values())`.


````python
# Exemple Drone Rescue — grille 12 colonnes, parcours et coordonnées 0-based
LETTRES = list('ABCDEFGHIJKL')
TAILLE  = 12

# Afficher les étiquettes de la 1ère ligne
for col in range(TAILLE):
    print(f'{LETTRES[col]}1', end='  ')
print()

# Compter les survivants en attente
survivants = {
    'S1': {'id': 'S1', 'etat': 'sauve'},
    'S2': {'id': 'S2', 'etat': 'en_attente'},
    'S3': {'id': 'S3', 'etat': 'en_attente'},
}
restants = sum(1 for s in survivants.values() if s['etat'] != 'sauve')
print(f'{restants} survivants restants')
print(f"Partie gagnée ? {all(s['etat'] == 'sauve' for s in survivants.values())}")
````
