## Concepts clés

### if / elif / else
```python
if batterie == 0:
    print('Drone HS')
elif batterie < 5:
    print('Batterie faible')
else:
    print('Drone opérationnel')
```

### for sur une liste
```python
for drone in drones:
    print(drone['id'])

# Avec index : enumerate()
for i, drone in enumerate(drones):
    print(i, drone['id'])
```

### while avec break
```python
while True:
    commande = input('> ').strip().upper()
    if commande in ('NEXT', 'QUIT'):
        break
    print('Commande non reconnue')
```

### any() / all()
```python
tous_sauves = all(s['sauve'] for s in survivants)
un_hs       = any(d['batterie'] == 0 for d in drones)
```


````python
# Exemples exécutables
drones = [
    {'id': 'D1', 'batterie': 10},
    {'id': 'D2', 'batterie': 0},
    {'id': 'D3', 'batterie': 3},
]

# for + if
for d in drones:
    if d['batterie'] == 0:
        print(f'{d["id"]} : HS')
    elif d['batterie'] < 5:
        print(f'{d["id"]} : batterie faible ({d["batterie"]})')
    else:
        print(f'{d["id"]} : OK ({d["batterie"]})')

# any() / all()
print(f'Un HS ? {any(d["batterie"]==0 for d in drones)}')
print(f'Tous OK ? {all(d["batterie"]>0 for d in drones)}')
````

La **boucle principale du jeu** est un `while True` qui tourne tant que la partie
