### Coordonnées 0-based et conversion affichage
```python
LETTRES = list('ABCDEFGHIJKL')

# (col=1, lig=2) → affichage 'B3'
def pos_str(col: int, lig: int) -> str:
    return f"{LETTRES[col]}{lig + 1}"

# 'B3' → (col=1, lig=2)
def pos_depuis_chaine(texte: str) -> tuple:

    lig = int(texte[1:]) - 1

```


````python
LETTRES = list('ABCDEFGHIJKL')

def pos_str(col: int, lig: int) -> str:
    """Convertit (col, lig) 0-based en notation affichage 'B3'."""
    return f"{LETTRES[col]}{lig + 1}"

def pos_depuis_chaine(texte: str) -> tuple:
    """Convertit 'B3' en (col=1, lig=2). Pas de validation ici."""
    texte = texte.strip().upper()
    col = LETTRES.index(texte[0])
    lig = int(texte[1:]) - 1
    return col, lig

# Tests
print(pos_str(0, 0))           # A1
print(pos_str(1, 2))           # B3
print(pos_depuis_chaine('B3')) # (1, 2)
print(pos_depuis_chaine('A1')) # (0, 0)
````

## Lien avec Drone Rescue

La **distance de Chebyshev** est la règle de mouvement centrale du jeu :
un drone ne peut se déplacer que d'une case (distance ≤ 1), diagonale incluse.
Elle travaille directement sur les entiers `col` et `lig` (0-based).


````python
def distance_chebyshev(col1: int, lig1: int, col2: int, lig2: int) -> int:
    """Distance de Chebyshev — diagonal = 1 pas. Coordonnées 0-based."""
    return max(abs(col2 - col1), abs(lig2 - lig1))

def case_valide(col: int, lig: int, taille: int = 12) -> bool:
    """True si (col, lig) est dans la grille (0-based)."""
    return 0 <= col < taille and 0 <= lig < taille

# Tests
print(distance_chebyshev(0, 0, 1, 1))   # 1 — diagonal OK
print(distance_chebyshev(0, 0, 2, 0))   # 2 — trop loin
print(case_valide(0, 0))                # True
print(case_valide(12, 0))               # False (hors grille 0-11)
````
