
````python
# Exemples exécutables — ord() et chr()
print(ord('A'))                   # 65
print(ord('L') - ord('A'))        # 11 → L est la 12e colonne (index 11)
print(chr(ord('A') + 5))          # F

# Construire la liste des colonnes valides
colonnes = [chr(ord('A') + i) for i in range(12)]
print(colonnes)  # ['A', 'B', ..., 'L']
````

## Lien avec Drone Rescue

La **distance de Chebyshev** est la règle de mouvement centrale du jeu :
un drone ne peut se déplacer que d'une case (distance ≤ 1), diagonale incluse.
Elle utilise `ord()` pour comparer les colonnes en lettres.


````python
def distance_chebyshev(col1: str, lig1: int, col2: str, lig2: int) -> int:
    """Distance de Chebyshev — diagonal = 1 pas."""
    dc = abs(ord(col1) - ord(col2))
    dl = abs(lig1 - lig2)
    return max(dc, dl)

def coord_valide(colonne: str, ligne: int, taille: int = 12) -> bool:
    """True si la coordonnée est dans la grille."""
    cols = [chr(ord('A') + i) for i in range(taille)]
    return colonne in cols and 1 <= ligne <= taille

# Tests
print(distance_chebyshev('A', 1, 'B', 2))   # 1 — diagonal
print(distance_chebyshev('A', 1, 'C', 1))   # 2 — trop loin
print(coord_valide('A', 1))                  # True
print(coord_valide('M', 1))                  # False
````
