## Lien avec Drone Rescue

Le module `jeu/logger.py` utilise `open()` + `with` pour écrire le fichier `.log`.
Le module `jeu/config.py` charge les paramètres depuis `config.json`
pour permettre des parties personnalisées (taille de grille, nombre de drones...).

**Format de sauvegarde d'une position (0-based) :**
```python
# Stocker : entiers 0-based
etat = {'tour': 3, 'score': 2, 'drones': [{'id': 'D1', 'col': 2, 'lig': 3}]}
# Afficher : LETTRES[col] + str(lig+1)
```


````python
import json
from datetime import datetime

LETTRES = list('ABCDEFGHIJKL')

def ecrire_log(chemin: str, messages: list):
    """Écrit les messages dans un fichier log avec horodatage."""
    with open(chemin, 'w', encoding='utf-8') as f:
        for msg in messages:
            horodatage = datetime.now().strftime('%H:%M:%S')
            f.write(f'[{horodatage}] {msg}\n')

# Sauvegarde d'un état (positions en 0-based)
etat = {'tour': 3, 'score': 2, 'drones': [{'id': 'D1', 'col': 2, 'lig': 3}]}
sauvegarde = json.dumps(etat, indent=2, ensure_ascii=False)
print(sauvegarde)

# Relire et afficher la position
etat_recharge = json.loads(sauvegarde)
d = etat_recharge['drones'][0]
print(f"D1 rechargé en {LETTRES[d['col']]}{d['lig']+1}")  # C4
````
