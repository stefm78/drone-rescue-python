## Concepts clés

### Définir une classe
```python
class Drone:
    def __init__(self, identifiant: str, colonne: str, ligne: int):
        self.identifiant = identifiant
        self.colonne = colonne   # 'A'..'L'
        self.ligne   = ligne     # 1..12
        self.batterie = 10

    def __str__(self) -> str:
        return f'{self.identifiant}@{self.colonne}{self.ligne}'
```

### @property
```python
@property
def est_hs(self) -> bool:
    return self.batterie <= 0   # recalculé à chaque accès

# Appel sans parenthèses :
if d.est_hs:   # ✅
    ...
```

### Héritage
```python
class EntiteGrille:
    def __init__(self, identifiant, colonne, ligne):
        ...

class Drone(EntiteGrille):          # hérite de EntiteGrille
    def __init__(self, identifiant, colonne, ligne):
        super().__init__(identifiant, colonne, ligne)  # OBLIGATOIRE
        self.batterie = 10
```


````python
# Classe de base EntiteGrille
class EntiteGrille:
    def __init__(self, identifiant: str, colonne: str, ligne: int):
        self.identifiant = identifiant
        self.colonne = colonne
        self.ligne   = ligne

    @property
    def position_str(self) -> str:
        return f'{self.colonne}{self.ligne}'

    def __str__(self) -> str:
        return f'[{self.identifiant}] {self.position_str}'

# Test EntiteGrille
e = EntiteGrille('X1', 'D', 5)
print(e)              # [X1] D5
print(e.position_str) # D5 — @property, pas de parenthèses !
````

## Lien avec Drone Rescue

Dans `jeu/modeles.py`, la classe `Position` gère les coordonnées en interne (int 0-based)
et expose une représentation str via `__str__`. Dans les **cours et exercices**,
on utilise directement `colonne: str` et `ligne: int` (convention pédagogique).


````python
# Classe Drone héritant de EntiteGrille
class Drone(EntiteGrille):
    def __init__(self, identifiant: str, colonne: str, ligne: int,
                 batterie_init: int = 10, batterie_max: int = 20):
        super().__init__(identifiant, colonne, ligne)
        self.batterie     = batterie_init
        self.batterie_max = batterie_max
        self.survivant    = None
        self.bloque       = 0

    @property
    def est_hs(self) -> bool:
        return self.batterie <= 0

    @property
    def est_bloque(self) -> bool:
        return self.bloque > 0

    def consommer_batterie(self):
        self.batterie = max(0, self.batterie - 1)

    def __str__(self) -> str:
        etat = 'HS' if self.est_hs else ('bloqué' if self.est_bloque else 'actif')
        return f'{self.identifiant} | {self.position_str} | bat:{self.batterie}/{self.batterie_max} | {etat}'

# Tests
d = Drone('D1', 'B', 7)
print(d)             # D1 | B7 | bat:10/20 | actif
print(d.est_hs)      # False
d.consommer_batterie()
print(d.batterie)    # 9
````
