## Concepts clés

### Dictionnaire imbriqué
```python
drone = {
    'id'          : 'D1',
    'col'         : 2,
    'lig'         : 3,
    'batterie'    : 10,
    'batterie_max': 20,
    'survivant'   : None,
    'bloque'      : 0,
    'hors_service': False,
}
```

### Dict de dicts — registre de tous les drones
```python
drones = {
    'D1': {'id': 'D1', 'col': 0, 'lig': 0, 'batterie': 10, ...},
    'D2': {'id': 'D2', 'col': 3, 'lig': 5, 'batterie': 20, ...},
}
# Accéder à D1
print(drones['D1']['batterie'])  # 10
```

### Set — ensemble de positions uniques
```python
zones_x = {(1, 2), (3, 4), (5, 5)}
zones_x.add((2, 2))          # ajouter
(1, 2) in zones_x            # True — test O(1)
```

### Parcourir un dict
```python
for did, drone in drones.items():
    print(did, drone['batterie'])

# Vérifier une condition sur tous les drones
tous_hs = all(d['hors_service'] for d in drones.values())
```


````python
# Créer un drone avec une fonction fabriquante
def creer_drone(identifiant, col, lig, batterie_max=20):
    """Retourne un dictionnaire représentant un drone."""
    return {
        'id'          : identifiant,
        'col'         : col,
        'lig'         : lig,
        'batterie'    : batterie_max // 2,
        'batterie_max': batterie_max,
        'survivant'   : None,
        'bloque'      : 0,
        'hors_service': False,
    }

# Créer 3 drones
drones = {
    'D1': creer_drone('D1', 0, 0),
    'D2': creer_drone('D2', 3, 2),
    'D3': creer_drone('D3', 7, 8),
}

# Afficher toutes les batteries
for did, drone in drones.items():
    print(f"{did} : batterie {drone['batterie']}/{drone['batterie_max']}")
````

## Lien avec Drone Rescue

Dans `jeu/logique.py`, toutes les entités du jeu sont des dictionnaires créés par des fonctions fabriquantes :
- `creer_drone(identifiant, col, lig)` → dict drone
- `creer_tempete(identifiant, col, lig)` → dict tempête
- `creer_survivant(identifiant, col, lig)` → dict survivant

L'état global du jeu est lui-même un dictionnaire : `etat['drones']`, `etat['survivants']`, `etat['zones_x']`...


````python
# Exemple inspiré du jeu : consommer de la batterie
COUT_TRANSPORT = 2  # coût si le drone transporte un survivant
COUT_ZONE_X    = 2  # coût supplémentaire zone dangereuse

def deplacer_drone(drone, cible, zones_x):
    """Déplace le drone vers cible et applique les coûts officiels."""
    cout = COUT_TRANSPORT if drone['survivant'] else 1
    if cible in zones_x:
        cout += COUT_ZONE_X
    drone['batterie'] = max(0, drone['batterie'] - cout)
    drone['col'], drone['lig'] = cible
    if drone['batterie'] == 0:
        drone['hors_service'] = True

# Test
d1 = creer_drone('D1', 0, 0)
zones = {(2, 2), (5, 3)}
deplacer_drone(d1, (1, 0), zones)   # déplacement normal
print(f"Batterie après déplacement normal : {d1['batterie']}")  # 9
deplacer_drone(d1, (2, 0), zones)   # pas en zone X
print(f"Batterie : {d1['batterie']}")  # 8
````
