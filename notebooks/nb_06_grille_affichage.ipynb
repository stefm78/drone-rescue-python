# Module 06 — Grille et affichage

## Objectifs
- Créer une grille ASCII avec une liste de listes
- Afficher des colonnes côte à côte avec `rjust` / `ljust`
- Convertir des coordonnées internes (int 0-based) en notation affichage (lettre + numéro)
- Calculer la distance de Chebyshev


## Concepts clés

### Grille comme liste de listes
```python
TAILLE = 5
grille = [['.' for _ in range(TAILLE)] for _ in range(TAILLE)]
grille[2][1] = 'D'   # ligne 2, colonne 1
```

### Notation affichage : lettre + numéro
```python
LETTRES = 'ABCDEFGHIJ'
# Case (col=1, lig=2) → affichage 'B3' (1-based)
def pos_str(col, lig):
    return f"{LETTRES[col]}{lig + 1}"
print(pos_str(1, 2))   # 'B3'
```

### Distance de Chebyshev
```python
# Maximum des différences absolues — autorise les diagonales
def distance_chebyshev(c1, l1, c2, l2):
    return max(abs(c2 - c1), abs(l2 - l1))

distance_chebyshev(0, 0, 1, 1)   # 1 (diagonale autorisée)
distance_chebyshev(0, 0, 2, 1)   # 2 (trop loin pour 1 déplacement)
```

### Alignement texte : rjust / ljust
```python
'D1'.ljust(4)    # 'D1  '  (complète à droite)
'10'.rjust(3)    # ' 10'   (complète à gauche)
f"{val:<6}"     # même chose avec f-string
```


````python
LETTRES = 'ABCDEFGHIJ'
TAILLE  = 5  # grille 5×5 pour cet exemple

def creer_grille(taille):
    """Retourne une grille vide de points."""
    return [['.' for _ in range(taille)] for _ in range(taille)]

def afficher_grille(grille):
    """Affiche la grille avec en-tête de colonnes et numéros de lignes."""
    taille = len(grille)
    print('      ' + '  '.join(LETTRES[:taille]))
    print('    ' + '---' * taille)
    for lig_idx, ligne in enumerate(grille):
        num = str(lig_idx + 1).rjust(2)
        cases = '  '.join(ligne)
        print(f"{num} |  {cases}")

# Test
g = creer_grille(TAILLE)
g[2][1] = 'D'   # drone en col=1, lig=2
g[0][4] = 'H'   # hôpital en col=4, lig=0
g[3][3] = 'S'   # survivant en col=3, lig=3
afficher_grille(g)
````

````python
def distance_chebyshev(c1, l1, c2, l2):
    """Distance de Chebyshev — autorise les déplacements diagonaux."""
    return max(abs(c2 - c1), abs(l2 - l1))

def pos_str(col, lig):
    """Convertit (col, lig) 0-based en notation affichage 'B3'."""
    return f"{LETTRES[col]}{lig + 1}"

# Tests distance de Chebyshev
assert distance_chebyshev(0, 0, 1, 1) == 1   # diagonale = 1 déplacement
assert distance_chebyshev(0, 0, 2, 1) == 2   # trop loin
assert distance_chebyshev(3, 3, 4, 4) == 1   # diagonale OK

# Tests notation
assert pos_str(0, 0) == 'A1'
assert pos_str(1, 2) == 'B3'
print('Tous les tests passent ✓')
````
