## Concepts clés

### Règles officielles du jeu
| Situation | Coût batterie |
|-----------|---------------|
| Déplacement normal | −1 |
| Transport d'un survivant | −2 |
| Entrée en zone X (supplément) | −2 |
| Recharge à l'hôpital | +3 |

### all() — vérifier une condition sur tous les éléments
```python

# Défaite : tous les drones sont hors service

```


````python
# Constantes officielles
COUT_TRANSPORT   = 2
COUT_ZONE_X      = 2
RECHARGE_HOPITAL = 3
GRILLE_TAILLE    = 5

def distance_chebyshev(c1, l1, c2, l2):
    return max(abs(c2 - c1), abs(l2 - l1))

def case_valide(pos):
    col, lig = pos
    return 0 <= col < GRILLE_TAILLE and 0 <= lig < GRILLE_TAILLE

def calculer_cout(drone, cible, zones_x):
    """Calcule le coût de batterie d'un déplacement selon les règles officielles."""
    cout = COUT_TRANSPORT if drone['survivant'] else 1
    if cible in zones_x:
        cout += COUT_ZONE_X
    return cout

# Tests
drone_vide    = {'id': 'D1', 'survivant': None, 'batterie': 10}
drone_chargé  = {'id': 'D2', 'survivant': 'S1', 'batterie': 10}
zones         = {(2, 2)}

assert calculer_cout(drone_vide,   (1, 1), zones) == 1  # normal
assert calculer_cout(drone_chargé, (1, 1), zones) == 2  # transport
assert calculer_cout(drone_vide,   (2, 2), zones) == 3  # normal + zone X
assert calculer_cout(drone_chargé, (2, 2), zones) == 4  # transport + zone X
print('Coûts OK ✓')
````
